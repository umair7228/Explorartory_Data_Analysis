### Basic plots with Matplotlib

#### line plot

##### defining X & Y axis
year = [1950, 1970, 1990, 2010] <br>
pop = [2.519, 3.692, 5.263, 6.972]

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating a line plot
year = [1880, 1900, 1920, 1940]
pop = [2.519, 3.692, 5.263, 6.972]

#Customizeing the size of line plot
figure = plt.figure()
figure.set_figwidth(4)
figure.set_figheight(3)
#making a plot
plt.plot(year, pop)
plt.show()
```

### Scatter plot

```python
#Customizing size
figure = plt.figure()
figure.set_figwidth(4)
figure.figure.set_figheight(3)

#making plot
plt.scatter(year, pop)
plt.show()
```

#### Histogram
- Explore dataset
- Get an idea about distribution
- `help()`
- [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
- `plt.hist(values, bins=3)`

```python
# Creating a Histogram
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]

#Customizing size
figure = plt.figure()
figure.set_figwidth(4)
figure.figure.set_figheight(3)

#making histogram
plt.hist(values, bins=5)
plt.show()
```
