# Case Study: Data Visualization (gapminder)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./gapminder.csv')
df.head(7)
```

```python
life_exp = df.life_exp
gdp_cap = df.gdp_cap

#Customizing Size of plot
figure = plt.figure()
figure.set_figwidth(4)
figure.set_figheight(3)

#Basic ploting and log scaling
plt.scatter(gdp_cap, life_exp)
plt.xscale('log')

xlab = 'GDP Per Capita [In USD]'
ylab = 'Life Expectancy [In Years]'
title = 'World Development in 2007'

#Labeling on Scatter plot
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

# tick values
tick_val = [1000, 10000, 100000]
tick_lab = ['1K', '10K', '100K']

#Adapt the ticks on x-axis
plt.xticks(tick_val, tick_lab)
plt.show()
```
