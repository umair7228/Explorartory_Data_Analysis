## Scenario: Analyzing and Visualizing Sales Data
### Objective
You are provided with a dataset containing monthly sales data for two different products over a year. Your task is to:

- Process and analyze the data using NumPy.
- Visualize the data using Matplotlib to identify trends and patterns.
#### Steps
##### 1. Generate the Data:
- Create a NumPy array with random sales data for two products over 12 months.
##### 2. Analyze the Data:
- Compute basic statistics (mean, sum, standard deviation) for each product.
##### 3. Visualize the Data:
- Plot the monthly sales data for both products using line plots.
- Add titles, labels, and a legend to the plot.
- Highlight the month with the highest sales for each product using scatter plots.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
months = np.arange(1, 13)
product1_sales = np.round(np.random.randint(10, 100, size=12) ,2)
product2_sales = np.round(np.random.randint(20, 120, size=12) ,2)

#Finding mean, sum and standard deviation
# Mean
mean_of_product1 = np.round(np.mean(product1_sales), 2)
mean_of_product2 = np.round(np.mean(product2_sales), 2)

#Sum
sum_of_product1 = np.sum(product1_sales)
sum_of_product2 = np.sum(product2_sales)

#Standard Deviation
std_of_product1 = np.round(np.std(product1_sales), 2)
std_of_product2 = np.round(np.std(product2_sales), 2)

print(f'Product 1 - Mean: {mean_of_product1}, Sum: {sum_of_product1}, Standard Deviation: {std_of_product1}')
print(f'Product 2 - Mean: {mean_of_product2}, Sum: {sum_of_product2}, Standard Deviation: {std_of_product2}')

#Ploting The Monthly Sales for both products
#Product 1
#Customize figure size   #87, 118
plt.figure(figsize=(10, 6))
#Ploting for both products
plt.plot(months, product1_sales, marker='o', label='Product 1 Sales', color='Blue')
plt.plot(months, product2_sales, marker='o', label='Product 2 Sales', color='Green')

#Calculating the maximum salse month for product 1
max_month_product1 = months[np.argmax(product1_sales)]
max_sales_product1 = np.max(product1_sales)
plt.scatter(max_month_product1, max_sales_product1, color='Red')
plt.text(max_month_product1, max_sales_product1, f'Max: {max_sales_product1}', color='Red', ha='left')

#Calculating the maximum salse moth for product 3
max_month_product2 = months[np.argmax(product2_sales)]
max_sales_product2 = np.max(product2_sales)
plt.scatter(max_month_product2, max_sales_product2, color='Red')
plt.text(max_month_product2, max_sales_product2, f'Max: {max_sales_product2}', color='Red', ha='left')

#Adding title, labels and legend
plt.title('Monthly Salse Data for Product 1 & Product 2')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.xticks(months)

plt.show()
```
