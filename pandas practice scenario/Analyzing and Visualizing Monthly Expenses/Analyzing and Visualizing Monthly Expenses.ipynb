# Scenario: Analyzing and Visualizing Monthly Expenses
#### Objective:
You are provided with a dataset containing monthly expenses for different categories over a year. Your task is to:

- Process and analyze the data using NumPy.
- Visualize the data using Matplotlib to identify trends and patterns.
#### Steps
1. Generate the Data:
- Create a NumPy array with random monthly expenses for three categories (e.g., Food, Entertainment, Utilities) over 12 months.
2. Analyze the Data:
- Compute basic statistics (mean, sum, standard deviation) for each category.
- Identify the month with the highest and lowest expenses for each category.
3. Visualize the Data:
- Plot the monthly expenses for each category using line plots.
- Use a stacked bar plot to show the total monthly expenses across all categories.
- Add titles, labels, and a legend to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate data
months = np.arange(1, 13)
categories = ['Food', 'Entertainment', 'Utilities']
expenses = np.round(np.random.randint(100, 500, size=(12, 3)), 2)  # 12 months, 3 categories

# Print expenses data
print("Expenses Data (Food, Entertainment, Utilities):")
print(expenses)

# Finding mean, sum, and standard deviation for each category
# Mean
mean_expenses = np.round(np.mean(expenses, axis=0), 2)
# Sum
sum_expenses = np.sum(expenses, axis=0)
# Standard Deviation
std_expenses = np.round(np.std(expenses, axis=0), 2)

print(f'Mean Expenses: {mean_expenses}')
print(f'Sum of Expenses: {sum_expenses}')
print(f'Standard Deviation of Expenses: {std_expenses}')

# Plotting the monthly expenses for each category
plt.figure(figsize=(12, 6))

# Plot for each category
for i, category in enumerate(categories):
    plt.plot(months, expenses[:, i], marker='o', label=f'{category} Expenses')

# Highlight the month with the highest expense for each category
for i, category in enumerate(categories):
    max_month = months[np.argmax(expenses[:, i])]
    max_expense = np.max(expenses[:, i])
    plt.scatter(max_month, max_expense, color='red', zorder=5)
    plt.text(max_month, max_expense, f'Max: {max_expense}', color='red', ha='left', zorder=5)

# Adding title, labels, and legend
plt.title('Monthly Expenses for Food, Entertainment, and Utilities')
plt.xlabel('Month')
plt.ylabel('Expenses (in $)')
plt.xticks(months)
plt.legend()

plt.grid(True)
plt.show()

# Stacked bar plot for total monthly expenses
total_expenses = np.sum(expenses, axis=1)

plt.figure(figsize=(12, 6))
plt.bar(months, expenses[:, 0], label='Food', color='blue')
plt.bar(months, expenses[:, 1], bottom=expenses[:, 0], label='Entertainment', color='green')
plt.bar(months, expenses[:, 2], bottom=expenses[:, 0] + expenses[:, 1], label='Utilities', color='orange')

# Adding title, labels, and legend
plt.title('Stacked Bar Plot of Monthly Expenses')
plt.xlabel('Month')
plt.ylabel('Expenses (in $)')
plt.xticks(months)
plt.legend()

plt.grid(True)
plt.show()
```
